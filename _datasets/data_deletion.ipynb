```python
#LIBRARIES TO BE USED

import pandas as pd
import numpy as np
```

```python
#READ DATASET (CSV FILE)

file_path = "google_trend_pinhao_original.csv"
df = pd.read_csv(file_path, skiprows=1, header=None, names=["Mes", "Pinhao: (Parana)"])

df.info()
df
```

```python
#PREPROCESS DATASET

df["Pinhao: (Parana)"] = pd.to_numeric(df["Pinhao: (Parana)"], errors='coerce') #Change from object to int64
df["Mes"] = pd.to_datetime(df["Mes"], format="%Y-%m", errors='coerce') #Change from string to date format
df = df.drop(0) #delete line zero due it is not relevant

df.info()
df
```

```python
#DELETE 10% OF DATA
#FUNCTION TO IDENTIFY PRIME NUMBERS THAT WILL BE USED TO DELETE DATA FROM PRIME NUMBER LINES
def primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

#CREATE A NEW DATAFRAME WITH 10% OF MISSING DATA
df_90 = df.copy() #Copy original dataframe
num_instances_to_delete = int(len(df) * 0.1) #Specify a limit of 10% to be deleted
deleted_instances = 0 #Count the quantity of data deleted to compare with limit above

#FOR FUNCTION THAT WILL DELETE THE DATA BACKWARDS, RESPECTING THE LIMIT
for index, row in reversed(list(df_90.iterrows())):
    if primo(index):
        df_90.loc[index, "Pinhao: (Parana)"] = pd.NA #Function that change the data for non-null to null/NA
        deleted_instances += 1
        if deleted_instances >= num_instances_to_delete:
            break

df_90.info()
df_90
```

```python
#DELETE 20% OF DATA
#FUNCTION TO IDENTIFY PRIME NUMBERS THAT WILL BE USED TO DELETE DATA FROM PRIME NUMBER LINES
def primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

#CREATE A NEW DATAFRAME WITH 10% OF MISSING DATA
df_80 = df.copy() #Copy original dataframe
num_instances_to_delete = int(len(df) * 0.2) #Specify a limit of 10% to be deleted
deleted_instances = 0 #Count the quantity of data deleted to compare with limit above

#FOR FUNCTION THAT WILL DELETE THE DATA BACKWARDS, RESPECTING THE LIMIT
for index, row in reversed(list(df_80.iterrows())):
    if primo(index):
        df_80.loc[index, "Pinhao: (Parana)"] = pd.NA #Function that change the data for non-null to null/NA
        deleted_instances += 1
        if deleted_instances >= num_instances_to_delete:
            break

df_80.info()
df_80
```

```python
# Exporta o dataframe df_90 para um arquivo CSV
df_90.to_csv('google_trend_pinhao_90percent.csv', index=False)
df_80.to_csv('google_trend_pinhao_80percent.csv', index=False)
```
